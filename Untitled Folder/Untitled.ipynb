```python
import numpy as np
import matplotlib.pyplot as plt
## Spectrum of a two-level system
def delta_noise(w,w_tls,amp):
    kappa = 1/(1e1) ## The total loss rate of a TLS (up + down)
    T_env = 0.6 ## in GHz, 0.6 GHz~30mK
    szeq = (np.exp(-w_tls/T_env) - 1)/(np.exp(-w_tls/T_env) + 1)
    return amp**2 *( (1+szeq)/2 * kappa/((w+w_tls)**2+(kappa/2)**2) +
                    (1-szeq)/2 * kappa/((w-w_tls)**2+(kappa/2)**2))  
def S2(w):
    noise=5e-3*w/(2*np.pi)
    w_tls =np.linspace(0,5,20)*2*np.pi## in GHz, tls frequency
    amp = 1.2e-1*np.random.rand(len(w_tls),1)/(2*np.pi) ## TLS coupling strength, should tune it manuallynp
    if w>=0:
        for j in range(len(w_tls)):
            noise=noise+ delta_noise(w,w_tls[j],amp[j])  
    if w<0:
        noise=0
    return noise
    
w_list = np.linspace(-5,5,1001)*2*np.pi
amplitude=np.zeros_like(w_list)
for i in range(len(w_list)):
    amplitude[i]=S2(w_list[i])
plt.plot(w_list/(2*np.pi), amplitude)
```
